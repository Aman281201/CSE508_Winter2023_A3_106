```python
import networkx as nx

Graph = nx.DiGraph()

with open("email-Eu-core.txt", "r") as file:     #data preprocessing
    data = file.read().splitlines()
all_nodes = set()
node_lst = list(all_nodes)
for line in data:
    u = 0
    v = 0
    u,v = line.split(" ")
    Graph.add_edge(u,v)
    all_nodes.add(u)
    all_nodes.add(v)
node_lst = list(all_nodes)
node_lst.sort()
for i in node_lst:
    Graph.add_node(i)

hubscores = nx.hits(Graph)[0]                    #calculates the hub score for all the nodes
authorityscores = nx.hits(Graph)[1]              #calculates the authority score for all the nodes


for n, hub_score in hubscores.items():
    a_score = authorityscores[n]
    print("Node" ,n, ": Hub score -", hub_score, ", Authority score -", a_score)
```
